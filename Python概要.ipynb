# 環境を導入
### Pythonをインストールしよう
#### Windows の場合
[Python for Windows インストールメモ](http://www.aoki.ecei.tohoku.ac.jp/~ito/python_windows.html)を見るとよい
* 公式サイトから最新版のインストーラをダウンロードしてインストール
* コマンドラインシェルで Python を使えるようにしたいので、インストールしたフォルダ (`C:\\Python\*\*` など) とその中の Scripts フォルダ (`C:\\Python\*\*\Scripts` など) のパスを通す
    - コマンドラインシェル: Windowsではコマンドプロンプトのこと。以下、シェルに打つことを単に「打つ」と書く 
    - パスを通す: シェルでどのディレクトリにいてもそのファイルを実行できるようにすること
        + Win8.1なら、スタートを右クリック→システム→システムの詳細設定→環境変数で、環境変数 PATH (なければ作る)にインストールしたフォルダと Scripts フォルダの場所を書き加えればよし。今のように要素が複数ある時は要素を;で区切って `C:\\Python\*\*;C:\\Python\*\*\Scripts` などと書く
        + パスが通っているかどうかは、`path` と打って全ての path を表示してみれば分かる

こうして、`python` と打つと Python 標準の対話型シェルが起動するようになる。
でも IPython の対話型シェル ipython の方が強力なので IPython パッケージを導入していこう

#### OS X の場合
もともと `python` と打つと起動するが、残念ながら Python 2。自分で Python 3 を入れよう
* 公式サイトから最新版のインストーラをダウンロードしてインストール

こうして、`python3` と打つと Python 3 が起動する


### パッケージ管理ソフト pip を使おう
Python の売りは豊富なパッケージ。
パッケージの一覧は [PyPI](https://pypi.python.org/pypi) (Python Package Index) で見られるけど、
ここからいちいち人間がパッケージをインストールしたりアップデートするのは大変。
そこで、パッケージ管理ソフトとよばれる「パッケージを自動で検索してインストールしたりアップデートしてくれるパッケージ」を使う。
* 「モジュール」とは再利用できるコードのファイルのことで、「パッケージ」とはたくさんのモジュールをディレクトリ構造にまとめて１つの再利用可能な部品のようにしたディレクトリのことっぽい

パッケージ管理ソフトにも幾つかあったが、今や pip が決定版。
昔は pip をダウンロードしてインストールするのに苦労してたみたいだけど、Python3.4 以降では元から pip が入っているので今や何もしなくてよい
* もはや詳しく知る必要がなくなったわけだけど、詳しくは[Python パッケージ管理技術まとめ (pip, setuptools, easy_install, etc)](http://www.yunabe.jp/docs/python_package_management.html) を参照

使い方は`pip`と打てば出る。fugaコマンドの使い方はヘルプオプション`-h`をつけて`pip fuga -h`と打てば出る。
[pipの使い方 (2014/1バージョン)](http://tdoc.info/blog/2014/01/15/pip.html)が詳しい

* Windowsでは、pip は既にパスを通してある Scripts フォルダに入ってるので、`pip` と打てばよい
* OS Xでは初め `pip3` と打つ必要があるが、`pip3 install -U pip` と打って最新版に更新したら `pip` が使える


#### よく使うものまとめ
* パッケージ hoge をインストール: `pip install hoge`
* パッケージ hoge をアップデート: `pip install -U hoge`
   - pip を最新版にするためにさっそく `pip install -U pip` と打ってみよう
* パッケージ hoge をアンインストール: `pip uninstall hoge`
* インストール済パッケージ一覧: `pip freeze` または `pip list` (こちらは pip と setuptools も表示される)
   - freezeでは、"パッケージ==バージョン"というフォーマット (要件フォーマット) で表示される。同じパッケージを全部インストールしたい他のコンピュータがあるときは、この一覧を txt ファイルにして渡し、要件オプション `-r` を使って `pip install -r hoge.txt` と打てばよい

### パッケージをインストールしよう
* IPython関連すべて: `pip install ipython[all]`

最近は `pip install jupyter` と打つほうが良いかも。 これで `ipython` と打てば IPython の対話型シェル ipython が起動するようになる。IPython Notebook を起動するには、 `ipython notebook` あるいは `jupyter notebook` と打つ

# コーディング規約
Pythonのコーディング規約としては、創始者Guidoがまとめた[PEP 8](https://www.python.org/dev/peps/pep-0008/)（[和訳](http://pep8-ja.readthedocs.org/ja/latest/)）が有名
* チェッカについては　http://qiita.com/ynakayama/items/8616f4c0c6e372de9a42
* PEP (Python Enhancement Proposal) は Python コミュニティにおける共有文章で、言語仕様の更新案や情報提供、ガイドラインなどが整理されている。各文章には固有の PEP 番号が割り振られていて、その一覧表が[PEP 0](https://www.python.org/dev/peps/)
    - リジェクトされた更新案も残っている。新しい提案を採用するか否かに対しては Python コミュニティの投票制らしい
* docstring (コード内での説明コメント) の書き方についての規約は、[PEP 257](https://www.python.org/dev/peps/pep-0257/): Docstring Convensions

Python 上のフレームワークのコーディング規約などといった他の規約もこの PEP 8 をベースにしている。詳しくは[Python の Coding Style あれこれ](http://tell-k.hatenablog.com/entry/2012/02/26/173554)
* Google による [Google Python Style Guide](http://google-styleguide.googlecode.com/svn/trunk/pyguide.html)（[和訳](http://works.surgo.jp/translation/pyguide.html)）は一読の価値あり
