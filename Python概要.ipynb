# 環境を導入しよう
### Pythonをインストールしよう
#### Windows の場合
[Python for Windows インストールメモ](http://www.aoki.ecei.tohoku.ac.jp/~ito/python_windows.html)を見るとよいよ
* 公式サイトから最新版のインストーラをダウンロードしてインストール
* コマンドラインシェルでPythonを使えるようにしたいので、インストールしたフォルダ(C:\\Python\*\*など)　とその中のScriptsフォルダ(C:\\Python\*\*\Scriptsなど)　のパスを通す
    - コマンドラインシェル: Windowsではコマンドプロンプトのこと。以下、シェルに打つことを単に「打つ」と書く 
    - パスを通す: シェルでどのディレクトリにいてもそのファイルを実行できるようにすること
        + Win8.1なら、スタートを右クリック→システム→システムの詳細設定→環境変数で、環境変数PATH(なければ作る)にインストールしたフォルダとScriptsフォルダの場所を書き加えればよし。今のように要素が複数ある時は要素を;で区切ってC:\\Python\*\*;C:\\Python\*\*\Scriptsなどと書く
        + パスが通っているかどうかは、`path`と打って全てのpathを表示してみれば分かる

こうして、`python`と打つとPython標準の対話型シェルが起動するようになる。
でもIPythonの対話型シェルipythonの方が強力なのでIPythonパッケージを導入していこう

#### Mac の場合
もともと`python`と打つと起動するが残念ながらPython 2。Python 3は入っていないので改めてPython 3を入れよう
* 公式サイトから最新版のインストーラをダウンロードしてインストール

こうして、`python3`と打つとPython 3が起動する


### パッケージ管理ソフト"pip"を使おう
Pythonの売りは豊富なパッケージ。
パッケージの一覧は[PyPI](https://pypi.python.org/pypi) (Python Package Index) で見られるけど、
ここからいちいち人間がパッケージをインストールやアップデートするのは大変。
そこで、パッケージ管理ソフトとよばれる「パッケージを自動で検索してインストールやアップデートしてくれるパッケージ」を使う。
* 「モジュール」とは再利用できるコードのファイルのことで、「パッケージ」とはたくさんのモジュールをディレクトリ構造にまとめて１つの再利用可能な部品のようにしたディレクトリのことっぽい

パッケージ管理ソフトにも幾つかあり、2015年1月時点での決定版がpip。
昔はダウンロードしてインストールするのに苦労してたみたいだけど、Python3.4以降には元からpipが入っているので今や何もしなくてよし
* もはや詳しく知る必要がなくなったわけだけど、詳しくは[Python パッケージ管理技術まとめ (pip, setuptools, easy_install, etc)](http://www.yunabe.jp/docs/python_package_management.html) を参照

使い方は`pip`と打てば出る。fugaコマンドの使い方はヘルプオプション`-h`をつけて`pip fuga -h`と打てば出る。
[pipの使い方 (2014/1バージョン)](http://tdoc.info/blog/2014/01/15/pip.html)が詳しい

* Windowsでは、pipは既にパスを通してあるScriptsフォルダに入ってるので、`pip`と打てばよい
* Macでは初め`pip3`と打つ必要があるが、`pip3 install -U pip`と打ち最新版に更新したら`pip`が使える


#### よく使うものまとめ
* パッケージhogeをインストール: `pip install hoge`
* パッケージhogeをアップデート: `pip install -U hoge`
   - pipを最新版にするためにさっそく　`pip install -U pip`してみよう
* パッケージhogeをアンインストール: `pip uninstall -U hoge`
* インストール済パッケージ一覧: `pip freeze` または `pip list` (こちらはpipとsetuptoolsも表示される)
   - freezeでは、"パッケージ==バージョン"というフォーマット (要件フォーマット) で表示される。同じパッケージを全部インストールしたい他のパソコンがあるときは、この一覧をtxtファイルにして渡し、要件オプション`-r`を使って`pip install -r hoge.txt`と打てばよい

### パッケージをインストールしよう
* IPython関連すべて: `pip install ipython[all]`

これで`ipython`と打てばIPythonの対話型シェルipythonが起動するようになる。IPython Notebookを起動するには`ipython notebook`と打つ
