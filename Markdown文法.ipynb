# 構造化する

### 段落わけ
Texと同様に、改行1回は1空白スペース扱い、改行2回以上ではじめて改行

### \#, \##, \###: 見出し
\#の数が多いほど小さい見出し。6個までできる

### ---: 水平線
\-が3つ以上と空白しかない行は水平線になる。\-は\*や\_でもよし

### \*, \-, \+, 0.: 箇条書き

行頭に4空白を入れるとネストできる。4空白追加するごとにネストが深まる

\*, \-, \+の3記号は区別されないが、 ネストのレベルごとに使い分けると自分が見やすい

0.のような数字とピリオドを使うと順序付になる。順序の記号は元の数字と無関係に自動で割り振られる

実例:

0. あ
 0. あい
    0. あいち
    0. あいす
 0. あお
    * あおこ
    * あおさ
0. い
 * いえ
    0. いえき
    0. いえん
 * いし
    * いしき
    * いしん

# 文章を装飾する

### \>：　引用
引用文中では行末2空白で改行する

\>\>でネストできる

> たとえば  
こんなかんじ
>> ネストも

> できる

### \`\`\`ではさむ: コード
(IPython NotebookではCodeセルがあるので不要)

\`\`\`ではさむ代わりに全体を4空白以上インデントしてもよい

\`\`\`の後にcやpythonなどの言語名を入れると、文法的機能に応じて文字が色わけされる（シンタックスハイライトという）

```c
#include <stdio.h>
int main(int argc, char **argv)
{
  printf("%s\n", "Hello, World");
  return 0;
}
```

```python
print("%s" %("Hello, World"))
```

インラインコードの場合は\`で囲む。たとえば`int main(void)`みたいなかんじ

### |: 図表

|で区切って
```
|a      |b   |c    |d     |
|-------|:---|----:|:----:|
|default|left|right|center|
|ee     |ff  |gg   |hh    |
```
などと書けば、

|a      |b   |c    |d     |
|-------|:---|----:|:----:|
|default|left|right|center|
|ee     |ff  |gg   |hh    |

と図表になる。2行目の:で各列の文字をどう揃えるかを指定する。
実は|の位置をそろえる必要はないし、両端の|も要らないので、
```
a|b|c|d
---|:--|--:|:-:
default|left|right|center
ee|ff|gg|hh
```
でもよい

# リンクを張る

### \<　\>: Webリンクを張る
\<http://hoge\>と書くと<http://google.com>のようにリンクが張られる

### \[　\]: 文字にリンクを張る
\[ぴよ\](http://hoge)あるいは\[\[ぴよ|http://hoge\]\] で文字「ぴよ」にリンクを張れる

\[ぴよ\](http://hoge "ほげ")とするとマウスオーバーで文字「ほげ」を出せる。
たとえば[こんな](http://google.com "ぐぐるよ")かんじ

参考文献のようにリンクを書くのを後回しにもできる。
いったん\[ぴよ\][piyo]とか書いといて、Markdownコードの一番下にでも

\[piyo\]: http://hoge "ほげ"

とか書けばよし

### \<a name= \>: 文字にアンカを張る
\#　\<a name ="fuga"\>ふが

で見出し「ふが」にアンカーがつく。文字「ぴよ」から「ふが」に飛べるようにするには、\[ぴよ\](#fuga)と書く。
たとえば[こんな](#1)かんじ

### \!\[　\]: 画像を表示
\[　\]の前に\!を付けるとリンク先の画像を表示できる。例:

![](http://python.fossee.in/static/website/images/pylogo.png)

\[　\]内の文字は無視されるっぽい

# 文字を入力する

### \\: 特殊文字のエスケープ
Markdown文法で使われる記号\#、\*などの特殊文字はそのまま書けないので、\\マークを前に付けて\\&、\\#などと書く

Markdownでは、HTML文法で使われる記号(&、<、>、")はそのまま書ける。
HTMLでは専用の面倒な書式(\&amp;、\&lt;、\&gt;、\&quot;)が必要だった

### \*, \*\*, \*\*\*ではさむ: 強調
1文字、2文字、3文字ではさむと、それぞれ_イタリック_、__太字__、___太字イタリック___

\*は\_でもよい

### \~~:　取り消し線 
たとえば~~こんな~~かんじ
