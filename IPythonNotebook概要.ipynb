IPython Notebookは、IPythonの「カーネル（計算実行部分）／フロントエンド（ユーザーインターフェース）」モデルにおける１つのフロントエンド。特徴：
* 従来のコンソール画面よりも自由度の高いMathematicaノートブック的なインターフェース
* Webアプリケーションである

コードの開発や実行、データ処理をブラウザ上で対話的に行いながら、入力コードや出力結果（matplotlibライブラリによる図など）からなる文章形式の*ノートブック*ファイル（.ipynbファイル）を作成できる
* LaTeXの数式やHTML、Markdown文法など多くの記述方法をサポート
* 内部はテキスト形式のJSON(JavaScript Object Notation)ファイルであり、IPythonをインストールせずに見られるし、バージョン管理が容易
* nbconvertコマンドにより、HTML, LaTeX, PDF, スライドショーなど様々な形式に変換できる
* ノートブック共有サービス[nbviewer](http://nbviewer.ipython.org/)で固有のウェブページとして公開・共有できる

# 使い方
* `ipython notebook`と打てばブラウザ上でダッシュボード画面が開く。ファイル一覧（ダッシュボード画面のFilesタブ）からノートブックを新規作成するか既存のノートブックを開く
    - 画面の一番上のロゴの右に出ているファイル名（デフォルトではUntitledとか）は直接編集できる
    - `.ipnb`ファイルのほか、`.py`ファイルも開ける
* ノートブックはカーネル（計算実行部分）に接続している。プロセス一覧（ダッシュボード画面のRunningタブ）でShutdownボタンを押さない限り、ノートブックを閉じてもカーネルはアクティブのまま。再びノートブックを開けばそのカーネルに再接続する

# セル
ノートブックは、*セル*と呼ばれる文章入力フィールドを並べたもの
* セルを選択して`Enter`でそのセル内を編集するモード（Editモード）へ入り、`Esc`でセルを選択するモード（Commandモード）へ戻る

セルは３種類
* Codeセル（Commandモードで`Y`）
    - コードを書く部分。デフォルトの言語はPython
    - セルを実行するとカーネルで計算され、計算結果がセルの出力として下に表示される。計算結果はテキストのみならず、図（matplotlibのプロット）やHTMLの表（pandasのデータ）などさまざま
* Markdownセル（Commandモードで`M`）
    - 文書や見出しを書く部分。HTMLやマークダウン文法、LaTeX(MathJax)が使える。$ではさむとインライン数式、$$ではさむと別行立ての数式が書けるほか、\\begin{equation}...\\end{equation}や\\begin{align}...\\end{align}といった標準的な数式環境も使えたり、\\newcommandでLaTeXマクロを定義できる
    - セルを実行するとmarkdwonやtexが実行されhtml文章に埋め込まれる
* Rawセル（Commandモードで`R`）
    - 出力として直接書くことができる。実行されない

`.ipnb`ではなく`.py`ファイルとして保存すると、セルの出力が消去されたりMarkdownセルがコメントアウトされたりする

セルの実行
* `Shift　Enter` セルを実行し、次のセルへ移動
* `Ctrl　Enter`　セルを実行するが、次のセルへ移動せずにそのセルにとどまる
* `Alt　Enter`　セルを実行し、直下に新しいセルを作って移動

# キーボードショートカット
マウスでも全ての操作ができるが、キーボードショートカットが便利。キーボードショートカットの一覧はCommandモードで`H`で見られる

* `Ctrl　S` : セーブする

Commandモード（セル選択モード）では
* `A` : 直上に新しいセルを作る
* `B` : 直下に新しいセルを作る
* `X` : セルをcut
* `C` : セルをcopy
* `V` : セルを直下にpaste
* `Shift　V` : セルを直上にpaste
* `DD` : セルを削除
* `Z` : 直前のセル削除を取り消す
* `Shift　M` : 下のセルと合併
* `L` : コードの行番号を表示／非表示
* `O` : 出力を表示／非表示
* `Space` : 下へスクロール
* `Shift　Space` : 上へスクロール
* `II` : カーネルを中断（interrupt）
* `00` : カーネルを再起動

Editモード（セル編集モード）では`Ctrl`を押しながら
* `]` : インデントする
* `[` : インデントを取り消す
* `A` : すべて選択
* `Z` : 直前の操作を取り消す
* `Y` : 取り消した操作をやり直す
* `↑/↓` : セルの一番上／下へ移動
* `→/←` : １単語ぶん右／左へ移動
* `Backspace` : 直前の１単語消去
* `Delete` : 直後の１単語消去
* `Shift　-` : セルを分割
