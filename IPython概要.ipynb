IPythonとは、もともとPythonにあった対話型計算環境を強力に拡張したもので、以下の３つが特徴
* ipython: もともとPythonにあった対話型シェルを強力に拡張した対話型シェル
* 「カーネル／フロントエンド」モデル: カーネル（計算実行部分）とフロントエンド（ユーザーインターフェース）の分離
    - カーネルを別のパソコン上に置いたり、１つのカーネルに複数の人が同時に接続したりできる
    - フロントエンドは数種類作られていて、その１つがウェブベースのIPython Notebook
* 並列計算お手軽設計

### タブ補完
何でも補完できる
* 変数、関数、オブジェクト、メソッドの名前
    - Pythonは完全なオブジェクト指向言語であり、変数、関数、メソッド、パッケージすべてがそもそもオブジェクトである
* インポートするときのパッケージの名前
* ファイル、ディレクトリの名前
* マジック関数(後述) の名前

### ヘルプ
hogeについて知りたければ`hoge?`あるいは`?hoge`と打つ
* Python標準シェルの`help()`関数より少し詳しい
* `hoge??`あるいは`??hoge`でより詳細な情報
* マジック関数(後述) にも使える

### マジック関数
IPython独自の対話型コマンド命令。その行1行を引数とするラインマジックと、続く何行かを引数とするセルマジックの２種類がある
* ラインマジックは`%hoge`、セルマジックは`%%hoge`という命名規則。ラインマジックの%は省略してもよい

例えばtimeitマジックは実行時間を計測できる。ある文piyoの実行時間を計測するには
```python
In [1]: %timeit piyo
```
あるいは
```python
In [1]: timeit piyo
```
と打つ。複数の文piyo1, piyo2, piyo3全体の実行時間を計測するには、
```python
In [1]: %%timeit piyo1
   ...: piyo2
   ...: piyo3
   ...:
```
と1文づつ入力して最後にEnterで打ち切る

マジック関数の種類は多すぎるのでよく使うものだけ覚えよう
* 各マジックの解説: [Built-in magic commands](http://ipython.org/ipython-doc/dev/interactive/magics.html)
* [よく使うマジック関数まとめ](マジック関数まとめ.ipynb)

### 履歴の参照
過去の入力がカーソルキーの上下↑↓でアクセスできるのみならず、過去の入出力はリスト`In`、`Out`に保存されている
* 直近３つの出力は`_`, `__`, `___`でもアクセスできる
* 過去の入力の一覧は`%hist`で見られる

### シェルコマンド
シェルコマンド`hoge`を使うには`!hoge`と打つ。`!ping google.com`など
* !!で結果を出力として取得
* シェルコマンドとはUNIXシェルコマンドの略であって、Windowsのシェル（コマンドプロンプト）ではもともとシェルコマンドが使えない。
    - WindowsでもUNIXコマンドが使える環境であるCygwinを導入しよう。Cygwinのシェルを使ってもよいが、Cygwinをインストールしたフォルダ(C:\cygwinとかC:\cygwin64とか)の下の\binのパスを通せばコマンドプロンプトでも使えるようになる
    - Windows上のUNIX環境としてはCygwinを使うよりは仮想マシンとしてLinuxを使うのが主流になってきたけれど、プログラミングで使う上では動作の速さを優先してCygwinを使う方が良いと思われる
        + 仮想マシン: OSを他のOS上の１アプリケーションとして動作させたもの。近年はマシンパワーの向上により簡単にできるようになった。VMwareやVirtualBoxが有名
        + 試しにVirtualBoxでUbuntuを使ってみたけどめっさ重い！！WindowsとLinuxの両方を使いたいだけならデュアルブートの方が快適だと感じたぞ

ある程度のシェルコマンドはラインマジックとして同様の機能のものが用意されており、これらは!無しで使えるしWindowsでもCygwin無しで使える。
ただ、標準的なシェルコマンドとは出力のフォーマットが少し違ったりする

* `pwd`　今いるディレクトリの場所を表示
* `ls`　ディレクトリ内のファイルの一覧を表示
* `cd`　ディレクトリの移動
* `pushd`/`popd`/`dirs`　実はディレクトリのスタックが用意されており、そのスタックを使うことで一時的に別のディレクトリに移動できる
    - `pushd` 今いるディレクトリを一時保存（スタックへ追加）してからディレクトリを移動
    - `popd` 最も最近に一時保存したディレクトリへ戻る（スタックからは削除される）
    - `dirs`　一時保存したディレクトリの一覧（スタック）を見る
* `mkdir`　ディレクトリの新規作成
* `rmdir`　ディレクトリの削除
* `copy`　ファイルのコピー
* `echo`　文字列の表示

なぜかラインマジックに無いっぽいもの
* `mv`　ファイルやディレクトリの移動、名前変更
* `rm`　ファイルの削除


`lst = !ls` のようにシェルスクリプトの出力結果を変数に保存したりできる
* コマンドプロンプトで試してみると日本語のファイルやフォルダ名が文字化けしてげんなり。たぶん、文字コードがUTF-8のCygwinの命令を文字コードがShift-JISのコマンドプロンプトから呼び出しているからなのだろう。今は世界的にUTF-8に統一する流れであるから、コマンドプロンプトは捨ててCygwinシェルを使うことにした。今度はpingが文字化けて使えない問題が発生したが…
    - Cygwinのホームディレクトリはデフォルトでは変なところにあるので、パスを通すときに環境変数PATHを定義したのと同様に環境変数HOMEを定義して好きなところに変えよう

## 感想
ipythonはとても便利
* Python標準の対話型シェルの上位互換だ

良いか悪いかは別として、IPythonはPythonの思想に完全には従っていないように見える
* 2系から3系になるときにprint文をprint()関数にしたように、Pythonは文法体系の簡潔さ・統一性を重んじている。
マジック関数は便利だけどその思想に逆行するように見える
* 「誰にとっても明確なやり方があるべきだ。そのやり方は唯一であることが好ましい」、というのがPythonの思想だ。同じ問題に対しては皆がほぼ同じコードを書き、個性は出ないけれどそのぶん可読性や保守性が高い。この思想こそが、"Perlのようなもの"仲間のRubyとの最も大きな違いだ。しかしIPythonでは、`%debug`と`%run -d`が同じであるなど、１つのことをするのに複数のやり方がある場合がしばしば

もしかしたら、IPythonは「エディタやIDEで書く文法を対話型シェルに強制するのは合理的ではない」という思想なのかもしれない
* 実際、対話型シェルで打つコードの多くはその場限りの使い捨てであり、他人も将来の自分も見ることのない、その瞬間の自分さえ分かっていれば良い部分だ。そういう部分なら可読性や保守性より利便性を優先してもよいのかもしれない
* というわけでマジック関数は便利だしギリギリ許せるけど、複数のやり方があるのを放置されてるのはちょっと嫌だな
