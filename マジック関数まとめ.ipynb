# ラインマジック
### 起動する
* `%run`　プログラム
    - `-t`　時間を計測する
    - `-d`　デバッガ
    - `-p` プロファイラ
* `%debug` デバッガ
* `%edit`　エディタ
* `%matplotlib`　matplotlib
* `%prun`　プロファイラ

* `%timeit`　かかる時間を計測

### 見る
* `%who`　現在定義されている変数名の一覧
* `%whos`　現在定義されている変数の一覧
* `%dirs`　現在定義されている辞書型の一覧
* `%who_ls`　現在定義されている変数のリスト

* `%page`　`print()`関数と同じ
* `%pdef`　その関数が何を引数に取るか（クラスに対してはコンストラクタの引数）
* `%pdoc`　そいつのdocstring
* `%pfile`　そいつが定義されてるファイル名
* `%psearch`　検索する
* `%pycat`
* `%hist`　入力の履歴を見る

### 設定する
* `%precision`　表示される数字の精度

### 開く
`%pylab`

### ログを残す
* `%logstart`　ログを取り始める
* `%logstop`　ログを取り終える
* `%save`　ファイルに保存する
* `%notebook` IPython notebookとして出力

* `%recall`　繰り返す
* `%reset`　リセットする

### その他
* `%lsmagic`　可能なマジックをすべて表示
    * たまにマジックの名前が昔と違ってたりマジックが削除されてたりするので
